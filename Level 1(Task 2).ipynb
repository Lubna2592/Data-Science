## Task:Descriptive Analysis
Calculate basic statistical measures (mean,median, standard deviation, etc.) for numericalcolumns.
Explore the distribution of categoricalvariables like "Country Code," "City," and"Cuisines."
Identify the top cuisines and cities with thehighest number of restaurants.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'Dataset .csv'
data = pd.read_csv(file_path)
```

```python
# Descriptive analysis for numerical columns
numerical_stats = data.describe()
print("Descriptive statistics for numerical columns:\n", numerical_stats)
```

```python
# Explore the distribution of categorical variables
# Distribution of 'Country Code'
plt.figure(figsize=(10, 6))
sns.countplot(x='Country Code', data=data, order=data['Country Code'].value_counts().index)
plt.title('Distribution of Country Code')
plt.xlabel('Country Code')
plt.ylabel('Count')
plt.show()
```

```python
# Plot distribution of 'City'
plt.figure(figsize=(12, 6))
sns.countplot(x='City', data=data, order=data['City'].value_counts().index)
plt.title('Distribution of Cities')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of 'Cuisines'
top_cuisines = data['Cuisines'].value_counts().head(10)
print("Top 10 Cuisines:\n", top_cuisines)
plt.figure(figsize=(14, 6))
sns.countplot(x='Cuisines', data=data, order=data['Cuisines'].value_counts().index)
plt.title('Distribution of Cuisines')
plt.xticks(rotation=90)
plt.show()
```

```python
#Identify Top Cuisines and Cities with the Highest Number of Restaurants:# Top cuisines
top_cuisines = data['Cuisines'].value_counts().head(10)
print("Top Cuisines:\n", top_cuisines)

top_cities = data['City'].value_counts().head(10)
print("Top 10 Cities with the Highest Number of Restaurants:\n", top_cities)
```
