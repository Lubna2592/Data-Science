## Task:Geospatial Analysis
Visualize the locations of restaurants on amap using latitude and longitudeinformation.
Analyze the distribution of restaurantsacross different cities or countries.
Determine if there is any correlationbetween the restaurant's location and itsrating.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'Dataset .csv'
data = pd.read_csv(file_path)
```

```python
# Create a scatter plot of restaurant locations
plt.figure(figsize=(10, 8))
plt.scatter(data['Longitude'], data['Latitude'], marker='o', s=30, alpha=0.5)
plt.title('Restaurant Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
```

```python
plt.figure(figsize=(12, 6))
data['City'].value_counts().plot(kind='bar')
plt.title('Distribution of Restaurants Across Cities')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Correlation Analysis
correlation = data[['Latitude', 'Longitude', 'Aggregate rating']].corr()
print('Correlation Matrix:')
print(correlation)

# Visualize correlation (example using a heatmap)
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
