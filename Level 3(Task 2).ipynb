## Task: Customer Preference Analysis
Analyze the relationship between the type of cuisine and the restaurant's rating. Identify the most popular cuisines among customers based on the number of votes. Determine if there are any specific cuisines that tend to receive higher ratings.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Data Collection
data = pd.read_csv('Dataset .csv')  
```

```python
# Step 2: Data Cleaning
data = data.dropna(subset=['Aggregate rating'])  # Drop rows with missing ratings
```

```python
# Step 3: Exploratory Data Analysis (EDA)
sns.scatterplot(x='Aggregate rating', y='Votes', data=data)
plt.title('Relationship between Rating and Votes')
plt.show()
```

```python
# Step 4: Identify Most Popular Cuisines
popular_cuisines = data.groupby('Cuisines')['Votes'].sum().sort_values(ascending=False).index
popular_cuisines
```

```python
# Step 5: Correlation Analysis
correlation_matrix = data[['Aggregate rating', 'Votes']].corr()
correlation_matrix
```

```python
# Step 6: Average Rating Analysis
average_ratings = data.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
average_ratings
```

```python
# Step 7: Visualization
sns.barplot(x=average_ratings.head(10).index, y=average_ratings.head(10).values)
plt.title('Average Rating by Cuisine (Top 10)')
plt.xticks(rotation=45)
plt.show()
```
