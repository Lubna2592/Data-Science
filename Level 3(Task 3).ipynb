## Task: Data Visualization
Create visualizations to represent the distribution of ratings using different charts (histogram, barplot, etc.).
Compare the average ratings of different cuisines or cities using appropriate visualizations.
Visualize the relationship between various features and the target variable to gain insights.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('Dataset .csv') 
```

```python
# Distribution of Ratings
plt.hist(data['Aggregate rating'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()
```

```python
# Box plot
sns.boxplot(x=data['Aggregate rating'])
plt.xlabel('Rating')
plt.title('Boxplot of Ratings')
plt.show()
```

```python
# KDE (Kernel Density Estimate) Plot
sns.kdeplot(data['Aggregate rating'], fill=True, color='skyblue')
plt.xlabel('Aggregate Rating')
plt.ylabel('Density')
plt.title('Distribution of Ratings (KDE)')
plt.show()
```

```python
# Compare Average Ratings of Different Cuisines
avg_rating_by_cuisine = data.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

plt.bar(avg_rating_by_cuisine.index, avg_rating_by_cuisine, color='lightgreen')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.title('Top 10 Cuisines by Average Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Compare Average Ratings of Different Cities
avg_rating_by_city = data.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

plt.bar(avg_rating_by_city.index, avg_rating_by_city, color='lightblue')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.title('Top 10 Cities by Average Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Visualize Relationship between Features and Target Variable
sns.scatterplot(x='Average Cost for two', y='Aggregate rating', data=data)
plt.xlabel('Average Cost for Two')
plt.ylabel('Aggregate Rating')
plt.title('Scatterplot of Average Cost vs. Aggregate Rating')
plt.show()
```

```python
# Assuming you want to visualize 'Average Cost for two', 'Votes', and 'Aggregate rating'
sns.pairplot(data[['Average Cost for two', 'Votes', 'Aggregate rating']])
plt.suptitle('Pairplot of Features vs. Aggregate Rating')
plt.show()
```

```python
#  Visualize Relationship Between Features and Target Variable
# Relationship between Average Cost for Two and Aggregate Rating
plt.scatter(data['Average Cost for two'], data['Aggregate rating'], alpha=0.5, color='orange')
plt.xlabel('Average Cost for Two')
plt.ylabel('Aggregate Rating')
plt.title('Relationship between Average Cost and Rating')
plt.show()

# Relationship between Votes and Aggregate Rating
plt.scatter(data['Votes'], data['Aggregate rating'], alpha=0.5, color='purple')
plt.xlabel('Votes')
plt.ylabel('Aggregate Rating')
plt.title('Relationship between Votes and Rating')
plt.show()
```
